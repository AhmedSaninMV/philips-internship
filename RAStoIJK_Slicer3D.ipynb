# RAS co-ordinate system to IJK co-ordinate system

This scipt can be used for converting 3D (x,y,z) co-ordinates in RAS system to IJK system.

The result (IJK co-ordinates) is saved in ".txt" file.
Input (RAS co-ordinates) shall be ".fcsv" only. (which is downloaded from 3D slicer after landmark annotation)

Results are saved in same folder as that of inputs.

Ensure that files names of ".nii" files and ".fcsv" files are same.

```python
import numpy as np
import pandas as pd
import glob
import nibabel as nib
```

```python
from os import listdir
landmark_files=[]
FileNames=listdir('/Users/ahmedsaninmv/Desktop/PHILIPS_INTERNSHIP/dataset/landmarks')
for names in FileNames:
    if names.endswith(".fcsv"):
        landmark_files.append(names)
```

```python
def rasMatrix(i,j,k):
    A = np.array([[i],[j],[k],[1]])
    return A
```

```python
for i in landmark_files:
    landmark_file = i 
    file_name = landmark_file.split('/')[-1].replace('.fcsv','')
    
#    print('file name', file_name,landmark_file)

    nii_file = [x for x in nii_files if file_name in x][0]
    img = nib.load(nii_file)
    affine_matrix = img.affine
    inverse_affine_matrix = np.linalg.inv(affine_matrix)
#    print(nii_file)
    
    landmark_path = open(landmark_file,'r')
    points = landmark_path.readlines()
    points_updated = points[3:]
    
    IJK_points = []
    save_points = []
    for k in points_updated:
    
        temp_ras = k.split(',')[1:4]
        temp_ras = list(map(float,temp_ras))
    
        temp_ijk = np.dot(inverse_affine_matrix,rasMatrix(temp_ras[0],temp_ras[1],temp_ras[2]))
        temp_ijk = list(map(int,temp_ijk))
        temp_ijk = np.transpose(temp_ijk).tolist()
        IJK_points.append(temp_ijk[:3])
    
        save_point = '{},{},{}\n'.format(temp_ijk[0],temp_ijk[1],temp_ijk[2])
        save_points.append(save_point)

    fsv_path = landmark_file.replace('.fcsv','.txt') 
    
                
    write = open(fsv_path, "w")
    write.writelines(save_points)
    write.close()
        

#     print('\n',RAS_points)    
#     print('\n',slicer_coordinates)
```
